### 2.5.1 Identifying potential custom stopwords not covered by NLTK's stopwords through Word Frequency 
<p style="font-size:16px"> I will be checking for potential stopwords through Word Frequency analysis. Although stopwords can already be removed by NLTK's stopwords, I am looking out for high frequency words that can be classified as a stopword that are not covered by NLTK's stopwords and adding it to a custom stopwords list later on.</p>

<p style="font-size:16px">The frequentWords function below, will remove NLTK stopwords and special character first, before returning the top 50 highest frequency words in each Rating Category's dataframe reviews.</p>

```python
from collections import Counter # Use to count the number of each word

def frequentWords(df, column='text'):
    # NLTK stopwords
    stop_words = set(stopwords.words('english'))

    # Combine all text from the specified column into a single string
    text = ' '.join(df[column].dropna().tolist())

    # Remove special characters from the text
    text = re.sub(r'[.,’"“”—–\-?;:()!\']', '', text)
    
    # Tokenize and filter out stopwords
    words = [word.lower() for word in word_tokenize(text) if word.lower() not in stop_words]
    
    # Count the frequency of each word
    word_counts = Counter(words)

    # Top 50 most common words
    return word_counts.most_common(50)

# Example usage
positive_top_words = frequentWords(unfiltered_positive)
neutral_top_words = frequentWords(unfiltered_neutral)
negative_top_words = frequentWords(unfiltered_negative)

print("Top 50 words in Positive reviews:", positive_top_words, "\n")
print("Top 50 words in Neutral reviews:", neutral_top_words, "\n")
print("Top 50 words in Negative reviews:", negative_top_words, "\n")
```

<b><p style="font-size: 16px"> Based on the top 50 highest frequency words from each rating category above & more, I have compiled the <b style="color: red">custom stopwords</b> list below.</p>

```python
custom = ['always', 'way','would', 'one', 'us', 'also', 'even', 'much', 'get', 'like', 'could', 'back', '2', 'got','told','said','flight','singapore','fly','good','go','back','told', 'didnt','never','dont','airline','airlines',"n’t"]
```

### 2.5.2 Removing stopwords with NLTK's stopwords and custom stopword list
<p style="font-size:16px"> The remove_stopwords function uses NLTK's stopword and the custom stopword list above to remove it from the review text. </p>

```python
def remove_stopwords(df, column='text'):
    # Combine NLTK stopwords with custom stopwords
    stop_words = set(stopwords.words('english')).union(set(custom))

    def process_text(text):
        if pd.isna(text):
            return text
        # Remove special characters and tokenize
        text = re.sub(r'[.,"“”—–\-?;:()!]', '', text)
        words = [word for word in word_tokenize(text.lower()) if word not in stop_words]
        return ' '.join(words)

    # Apply processing to the specified column
    df[column] = df[column].apply(process_text)
    return df
```

<h2>2.6 Lemmatization</h2>

<p style="font-size:16px"> <b>I have chosen to use lemmatization instead of stemming.</b></p> 

<p style="font-size:16px"> <b>Stemming</b> simply chops off suffixes to reach a root form, for example, it changes "dancing" to "danc" which is not a english word, while <b>lemmatization</b> produces more meaningful results by reducing words to their root form while preserving their meaning in context, for example, "dancing" will be converted to "dance" which is a valid english word. </p>

<p style="font-size:16px"> I will apply lemmatization on the filtered dataframe created above.

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_text(df, column='text', pos='v'):
    lemmatizer = WordNetLemmatizer()

    def process_text(text):
        if pd.isna(text):
            return text
        # Tokenize and lemmatize words
        words = word_tokenize(text)
        lemmatized_words = [lemmatizer.lemmatize(word, pos) for word in words]
        return ' '.join(lemmatized_words)

    # Apply lemmatization to the specified column
    df[column] = df[column].apply(process_text)
    return df
```

```python
filtered_positive = lemmatize_text(filtered_positive)
filtered_neutral = lemmatize_text(filtered_neutral)
filtered_negative = lemmatize_text(filtered_negative)
```

<h2>3.1 Sentiment Analyzing function</h2>

<p style="font-size:16px">The <b>analyze_sentiment</b> function uses a internal function called <b>sentiment_type</b> to go through every review in the dataframe, combines its text and title and uses <b>NLTK's VADER Sentiment Analysis Tool</b> to generate a compounded polarity score and returns a sentiment type. Afterwards, it adds the sentiment type to a new column called <b>"Sentiment Type"</b> in the dataframe.</p>

```python
# Instance of the nltk sentiment analyser
analyzer = SentimentIntensityAnalyzer() 

def analyze_sentiment(df):

    def sentiment_type(row):
        # Combine text and title
        combined_text = f"{row['title']} {row['text']}"
        polarity_score = analyzer.polarity_scores(combined_text)['compound']

        # Determine sentiment rating based on polarity score
        if polarity_score > 0.01:
            return "Positive"
        elif polarity_score > -0.01:
            return "Neutral"
        else:
            return "Negative"

    # Apply sentiment analysis to each review
    df['sentiment type'] = df.apply(sentiment_type, axis=1)
    
    # Reorder columns to place "Sentiment rating" next to "Rating Category"
    columns = list(df.columns)
    columns.insert(columns.index('rating category') + 1, columns.pop(columns.index('sentiment type')))
    df = df[columns]

    return df
```

<p style="font-size=16px">Note: The range of compounded polarity scores between 0.01 and -0.01 is suitable for a neutral rating because it reflects minimal sentiment, indicating that the text is neither strongly positive nor negative, but rather balanced or ambiguous.</p>

### 3.1.1 Applying analyze_sentiment function on each unfiltered rating category dataframe

```python
unfiltered_positive = analyze_sentiment(unfiltered_positive)
unfiltered_neutral = analyze_sentiment(unfiltered_neutral)
unfiltered_negative = analyze_sentiment(unfiltered_negative)

# Check if function creates new Sentiment Type column and assigns value
unfiltered_positive.head()
```

<h2>3.2 Match Percentage between Rating Category and Sentiment Type</h2>

<p style="font-size:16px">I will now check the amount of matches in each Rating Category and generate a matching percentage. Afterwards, I will use it to generate a bar graph as data visualization and attempt to find out if the sentiment analyzer's accuracy varies significantly across the different rating categories (Positive, Neutral, and Negative).</p>

```python
def sentiment_match(df):
    match_count = (df['rating category'] == df['sentiment type']).sum()
    total_count = len(df)
    return (match_count / total_count) * 100
```

```python
# Calculate match percentages
positive_match = sentiment_match(unfiltered_positive)
neutral_match = sentiment_match(unfiltered_neutral)
negative_match = sentiment_match(unfiltered_negative)
```

```python
# Bar Graph Data Visualization
categories = ['Positive', 'Neutral', 'Negative']
match_percentages = [positive_match, neutral_match, negative_match]

plt.figure(figsize=(8, 5))
plt.bar(categories, match_percentages, color=['green', 'orange', 'red'])
plt.title('Match Percentage Between Rating Category and Sentiment Type', fontsize=14)
plt.xlabel('Rating Category', fontsize=12)
plt.ylabel('Match Percentage (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', alpha=0.2)
plt.show()
```

<h2> 3.3 Analysis & Summary</h2>


### 3.3.1 Analysis based on the visualization above:

<p style="font-size:16px"><b style="color: green">Positive Reviews:</b> The sentiment analyzer demonstrates a high degree of accuracy for the "Positive" category, with nearly 100% of positive reviews correctly matching their sentiment rating.</p>

<p style="font-size:16px"> <b style="color: orange">Neutral Reviews:</b> The "Neutral" category has a very low match percentage, nearing 0%, suggesting that the sentiment analyzer struggles with distinguishing truly neutral sentiments. This could be due to the nature of neutral reviews, which often have subtle language that leans slightly positive or negative but isn't captured as neutral by the tool.</p>

<p style="font-size:16px"><b style="color: red">Negative Reviews:</b> The sentiment analyzer shows a decent match percentage for the "Negative" category (around 70%), but it is less accurate than for positive reviews. This discrepancy might be due to the variety of language and tone used in negative reviews, which can sometimes include mixed sentiments or sarcasm, making them harder for the tool to classify correctly.</p>

### 3.3.2 Analysis Conclusion
<p style="font-size:16px">We can see that for positive and negative reviews, the sentiment of the review text greatly aligns with the rating score that the user has given which has been categorized above in <a href="#2.3">Section 2.3.</a> However, neutral reviews are greatly misaligned between their sentimental type and given rating which could be because of sentences and words that lean more towards a positive or negative tone while the passenger naturally feels neutral towards Singapore Airlines.</p>

### 4.2.1 Review Statistic of each Rating Category (Positive, Neutral, Negative)
<p style="font-size:16px">I will generate key summary statistics for each rating category (Positive, Neutral, Negative) and visualize the results with graphs. These visualizations will help uncover trends in the data and provide deeper insights into the underlying reasons for these patterns. By analyzing these trends, we aim to better understand passenger behavior across different review categories. </p>

```python
def review_statistics(pos_df, neu_df, neg_df):
    # Calculate statistics for each category
    stats = {
        "Category": ["Positive", "Neutral", "Negative"],
        "Number of Reviews": [len(pos_df), len(neu_df), len(neg_df)],
        "Average Words per Review": [
            pos_df['text'].str.split().apply(len).sum() / len(pos_df),
            neu_df['text'].str.split().apply(len).sum() / len(neu_df),
            neg_df['text'].str.split().apply(len).sum() / len(neg_df),
        ]
    }
    
     # Convert to DataFrame for plotting
    stats_df = pd.DataFrame(stats).set_index("Category")
    
    # Colors for each category
    colors = ["green", "orange", "red"]
    
    # Plot settings
    plt.figure(figsize=(12, 10))
    for i, col in enumerate(stats_df.columns, 1):
        plt.subplot(2, 2, i)  # Arrange plots in a 2x2 grid
        
        # Positions for bar plot
        x_positions = range(len(stats_df[col]))
        
        # Create bar plot Matplotlib
        plt.bar(x_positions, stats_df[col], color=colors)
        
        # Add labels and titles
        plt.title(col)
        plt.ylabel(col)
        plt.xlabel("Rating Category")
        
        # Set x-tick labels to the rating categories
        plt.xticks(x_positions, stats_df.index)
    
    plt.tight_layout()
    plt.show()

# Example usage
review_statistics(filtered_positive, filtered_neutral, filtered_negative)
```

```python
# Ensure the 'published_date' column is in datetime format
filtered_positive['published_date'] = pd.to_datetime(filtered_positive['published_date'])
filtered_negative['published_date'] = pd.to_datetime(filtered_negative['published_date'])

# Create new dataframes and filter reviews by years for positive reviews (2018 and 2019)
positive_2018 = filtered_positive[filtered_positive['published_date'].dt.year == 2018]
positive_2019 = filtered_positive[filtered_positive['published_date'].dt.year == 2019]

# Create new dataframes and filter reviews by years for negative reviews (2018 and 2019)
negative_2018 = filtered_negative[filtered_negative['published_date'].dt.year == 2018]
negative_2019 = filtered_negative[filtered_negative['published_date'].dt.year == 2019]

years = ['2018', '2019'] # x-axis values
positive_counts = [len(positive_2018), len(positive_2019)]
negative_counts = [len(negative_2018), len(negative_2019)]

# Prepare data for stacked bar chart plot
data = pd.DataFrame({
    'Positive Reviews': positive_counts,
    'Negative Reviews': negative_counts
}, index=years)

data.plot(kind='bar', stacked=True, figsize=(8, 5), color=['green', 'red'])
plt.title('Number of Reviews (Positive vs Negative) in 2018 vs 2019', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Reviews', fontsize=12)
plt.legend(title='Review Type')
plt.grid(True)
plt.show()
```
